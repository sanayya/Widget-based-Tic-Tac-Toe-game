```python
from ipywidgets import Button, HBox, VBox, Label, Layout
from IPython.display import display, clear_output
from random import shuffle
from time import sleep
 
class TTT:
    
    winning_lines = [(0,1,2), (3,4,5), (6,7,8), (0,3,6), (1,4,7), (2,5,8), (0,4,8), (2,4,6)]           
    buttons = [Button(description = '') for _ in range(9)] # create 9 ipywidgets buttons
    
    def __init__(self): # create the game board and setup initial values for some variables
 
        for button in TTT.buttons:
            button.layout = Layout(width='50px', height='50px')
            button.style.button_color = '#bfedfb' # or 'lightblue' gives the same result
            button.on_click(self.button_clicked)
 
        first_row =  HBox([TTT.buttons[0], TTT.buttons[1], TTT.buttons[2]])
        second_row = HBox([TTT.buttons[3], TTT.buttons[4], TTT.buttons[5]])
        third_row =  HBox([TTT.buttons[6], TTT.buttons[7], TTT.buttons[8]])
        self.vbox = VBox([first_row, second_row, third_row])
        self.play_again_button = Button(description="Play again")
        self.play_again_button.on_click(self.play_again)
        self.play_again_button.layout.visibility = 'hidden'
        self.dont_play_again_button = Button(description="End Game")
        self.dont_play_again_button.on_click(self.end_game)
        self.dont_play_again_button.layout.visibility = 'hidden'
        self.hbox = HBox([self.play_again_button, self.dont_play_again_button])
        
        self.label = Label(value = '')
        display(self.label)
        self.players = []
        self.turn_index = 0
        self.game_over = False
        
        
        
    def button_clicked(self,b): # b is the button that was clicked
        player = self.players[self.turn_index]
        nextPlayer = self.players[1 - self.turn_index]
        if (b.description != 'X' and b.description != 'O') and not self.game_over :
            self.label.value = f"{nextPlayer.name}, place your {nextPlayer.marker}"
            b.description = player.marker
            self.end_check(player)
            if not self.game_over:
                self.turn_index = 1 - self.turn_index
            else:
                self.play_again_button.layout.visibility = 'visible'
                self.dont_play_again_button.layout.visibility = 'visible'
    
 
 
 
    def end_check(self,player):
        for a,b,c in self.winning_lines:
            marks = (TTT.buttons[a].description, TTT.buttons[b].description, TTT.buttons[c].description)
            if marks == (player.marker,player.marker,player.marker):
                self.label.value = f'{player.name} wins!'
                self.game_over = True
 
        if not '' in [button.description for button in TTT.buttons] and not self.game_over:
            self.label.value = 'It\'s a Tie. Game Over!'
            self.game_over = True
            
            
            
            
    def play_again(self, b):
        self.play_again_button.layout.visibility = 'hidden'
        self.dont_play_again_button.layout.visibility = 'hidden'
        for button in TTT.buttons:
            button.description = ''
        self.turn_index = 1 - self.turn_index 
        player = self.players[self.turn_index] # alternate which player starts the game
        self.label = Label(value = f"{player.name}, place your {player.marker}")
        clear_output()
        self.game_over = False
        display(self.label, self.vbox, self.hbox)
        
        
    def end_game(self, b):
        self.label = Label(value = 'Game over. Goodbye!')
        clear_output()
        display(self.label)    
            
 
class Player:
    def __init__(self, name, marker):
        self.name = name
        self.marker = marker
 
 
    
def start_game():
    
    board = TTT()
    print('Welcome to Tic Tac Toe!')
    sleep(0.2)
    players = [Player('One', 'none'), Player('Two', 'none')]
    players[0].name = input('Pick your name for this game: ')
    clear_output()
    players[1].name = input('Other player, pick your name for this game: ')
    clear_output()
    shuffle(players)
    print(f'{players[0].name}, you go first')
    chosen_marker = 'none'
    markers = ['X', 'O']
    while chosen_marker not in markers:
        chosen_marker = input(f"{players[0].name}, choose 'X' or 'O' as your marker:  ").upper()
    players[0].marker = chosen_marker
    markers.remove(chosen_marker)
    players[1].marker = markers[0]
        
    clear_output()
    board.label.value = f'{players[0].name}, Place your {players[0].marker}'  
    display(board.label, board.vbox, board.hbox)
    board.players = players
    
start_game()
```
